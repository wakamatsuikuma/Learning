### データフレーム
- 行と列の概念を持つ2次元のデータ構造
- 行もしくは列の参照には、[または$を使う
- [行、列]で参照する。入力しない場合は、全てが指定される。
- data.frame()関数で作成

```python
# data.frame()でデータフレーム作成
# 引数は、列名=値の形式　で指定する

dat <- data.frame(
  X = 1:5,
  Y = seq(2, 10, by=2),
  Z = c("a", "b", "c", "d", "e")
)

print(dat)
```

```python
print(dat[2, ])
print(dat[ , 2])
print(dat[1, 1])
```

```python
# 列名でも指定できるが文字列型なので、[なら""付き、$ならそのまま！
print(dat[, "X"])
print(dat$X)
```

### 行列
- データフレーム同様に行と列からなる2次元データであるが、全ての要素が同じデータ型でないといけない
- matrix()で生成。nrowとncolでサイズ指定する。
- 各要素に積をとる、内積計算、行または列の統計量をとる などができる
- 参照方法は、データフレーム同様に[でできる
- [で取得すると、ベクトルに変換されるときがあるが、drop=Falseを追記すれば避けられる

```python
m <- matrix(1:9, nrow=3, ncol=3)
print(m)
```

```python
# 各要素に対する積をとる
print(m * 3)
```

```python
# 内積計算
print(m %*% m)
```

```python
# 行ごとの総和
print(rowSums(m))
```

```python
# 列ごとの平均
print(colMeans(m))
```

```python
# ベクトルへの変換の回避
print(m[ , 3])
print(m[ , 3, drop=FALSE])
```
