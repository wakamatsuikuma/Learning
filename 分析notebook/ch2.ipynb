```python
x <- seq(0, 1, 0.01)
y <- 1 - x^4
plot(x, y)

cor.test(x, y)
cor.test(x, y, method = "spearman")
cor.test(x, y, method = "kendall")
```

```python
x <- seq(0, 1, 0.01)
y <- 1 - x^4
c1x <- seq(0, 1, 0.01)*2
c2y <- (1 - x^4)*3

cor(x, y)
cor(x, y, method = "spearman")
cor(x, y, method = "kendall")
cor(c1x, c2y)
cor(c1x, c2y, method = "spearman")
cor(c1x, c2y, method = "kendall")
```

```python
x <- seq(-1, 1, 0.01)
y <- x^3 - 3*x + rnorm(length(x), 0, 0.1)
cor(x, y)
plot(x, y)
```

```python
M <- 100
a <- 0
b <- 1
N <- 1000
v <- numeric(N)


for (i in 1:N){
x <- runif(M, a, b)
y <- runif(M, a, b)
tau <- cor(x, y, method="spearman")
rho <- cor(x, y, method="kendall")
v[i] <- 3*tau - 2*rho
}

plot(v, ylim=c(-1.5, 1.5))
abline(h=-1)
abline(h=1)
```

```python
x1 <- rnorm(100, 150, 10)
x2 <- rnorm(100, 150, 10)
plot(x1, x2)
print(cor(x1, x2))

y1 <- rnorm(100, 50, 5)
y2 <- rnorm(100, 50, 5)
plot(y1, y2)
print(cor(y1, y2))

xy1 <- c(x1, y1)
xy2 <- c(x2, y2)
plot(xy1, xy2)
print(cor(xy1, xy2))
```
