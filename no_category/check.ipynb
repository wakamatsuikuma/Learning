```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# 逆ロジット変換の概形確認
x = np.random.rand(200)*100
x = ( x - 50 ) / 10
y = np.exp(x) / (1 + np.exp(x))

plt.scatter(x, y)
```

```python
# 指数変換の概形確認
a = 1.01
x = np.arange(1, 101)
y = a ** x
plt.scatter(x, y)
```
